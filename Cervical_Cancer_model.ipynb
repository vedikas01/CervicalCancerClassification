```python
# full_script_with_kagglehub.py
# Full optimized training script using kagglehub to download the SIPaKMeD dataset

import os
import zipfile
import random
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# attempt to import kagglehub for dataset download
try:
    import kagglehub
except Exception as e:
    raise ImportError("kagglehub not installed. Install it with `pip install kagglehub` and re-run. Error: " + str(e))

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, regularizers
from tensorflow.keras.applications import MobileNetV2

# ---------------------------
# CONFIG
# ---------------------------
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS_FEATURE = 15
EPOCHS_FINETUNE = 15
MODEL_DIR = "./models"
os.makedirs(MODEL_DIR, exist_ok=True)
AUTOTUNE = tf.data.AUTOTUNE

# ---------------------------
# Download dataset using kagglehub
# ---------------------------
print("Downloading SIPaKMeD dataset via kagglehub...")
# this call should download and return local path (zip or folder) depending on kagglehub behavior
ds_path = kagglehub.dataset_download("prahladmehandiratta/cervical-cancer-largest-dataset-sipakmed")
print("kagglehub returned:", ds_path)

# ds_path may be a string path to a zip or folder, or a dict - handle common cases:
if isinstance(ds_path, (list, tuple)) and len(ds_path) > 0:
    ds_path = ds_path[0]
if isinstance(ds_path, dict) and "path" in ds_path:
    ds_path = ds_path["path"]
if not isinstance(ds_path, str):
    raise RuntimeError("Unexpected return from kagglehub.dataset_download — got: %r" % ds_path)

# extract if zip
if ds_path.lower().endswith(".zip"):
    extract_to = os.path.splitext(ds_path)[0]
    if not os.path.exists(extract_to):
        print(f"Extracting {ds_path} -> {extract_to} ...")
        with zipfile.ZipFile(ds_path, "r") as z:
            z.extractall(extract_to)
    BASE_DIR = extract_to
else:
    # if dataset is already a directory
    BASE_DIR = ds_path

print("Dataset base directory:", BASE_DIR)

# ---------------------------
# discover images & labels
# ---------------------------
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}
def is_image_file(fn): return os.path.splitext(fn.lower())[1] in IMG_EXTS

def discover_images_and_labels(root_dir):
    items = []
    for r, dirs, files in os.walk(root_dir):
        if r.endswith("__MACOSX") or r.endswith(".ipynb_checkpoints"): continue
        images = [f for f in files if is_image_file(f)]
        if not images: continue
        label = os.path.basename(r)
        if label.lower().startswith("._"): continue
        for im in images:
            items.append((os.path.join(r, im), label))
    return items

pairs = discover_images_and_labels(BASE_DIR)
if not pairs:
    raise RuntimeError(f"No images found under {BASE_DIR} — inspect dataset contents.")
print(f"Found {len(pairs)} images across {len(set([p[1] for p in pairs]))} labels")

# ---------------------------
# prepare splits and TF datasets
# ---------------------------
filepaths, labels = zip(*pairs)
filepaths = np.array(filepaths)
labels = np.array(labels)
unique_labels = sorted(list(set(labels)))
label_to_index = {l:i for i,l in enumerate(unique_labels)}
indices = np.array([label_to_index[l] for l in labels])

train_fp, temp_fp, y_train, y_temp = train_test_split(filepaths, indices, test_size=0.30, random_state=SEED, stratify=indices)
val_fp, test_fp, y_val, y_test = train_test_split(temp_fp, y_temp, test_size=0.50, random_state=SEED, stratify=y_temp)
print(f"Train: {len(train_fp)}  Val: {len(val_fp)}  Test: {len(test_fp)}")

num_classes = len(unique_labels)

def decode_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32) / 255.0
    return img, tf.one_hot(label, depth=num_classes)

augmenter = tf.keras.Sequential([
    layers.RandomFlip('horizontal_and_vertical'),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.18),
    layers.RandomContrast(0.15),
])

def make_ds(fps, ys, training=False):
    ds = tf.data.Dataset.from_tensor_slices((fps, ys))
    if training:
        ds = ds.shuffle(buffer_size=len(fps), seed=SEED)
    ds = ds.map(lambda p,y: decode_image(p,y), num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augmenter(x, training=True), y), num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_fp, y_train, training=True)
val_ds = make_ds(val_fp, y_val, training=False)
test_ds = make_ds(test_fp, y_test, training=False)

print("Datasets ready. Classes:", unique_labels)

# ---------------------------
# model definitions (CNN & Hybrid)
# ---------------------------
def build_mobilenetv2_cnn(input_shape, num_classes):
    base = MobileNetV2(include_top=False, weights='imagenet', input_shape=input_shape, pooling='avg')
    base.trainable = False
    inputs = layers.Input(shape=input_shape)
    x = base(inputs, training=False)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(0.001))(x)
    model = models.Model(inputs, outputs, name='mobilenetv2_improved')
    return model, base

def build_cnn_rnn_hybrid(input_shape, num_classes):
    base = MobileNetV2(include_top=False, weights='imagenet', input_shape=input_shape, pooling=None)
    base.trainable = False
    inputs = layers.Input(shape=input_shape)
    x = base(inputs, training=False)
    seq_len = int(x.shape[1] * x.shape[2])
    feat_dim = int(x.shape[3])
    x = layers.Reshape((seq_len, feat_dim))(x)
    x = layers.Bidirectional(layers.LSTM(256, return_sequences=True))(x)
    x = layers.Bidirectional(layers.LSTM(128))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(0.001))(x)
    model = models.Model(inputs, outputs, name='cnn_rnn_hybrid')
    return model, base

# ---------------------------
# training helpers
# ---------------------------
def compile_and_train(model, train_ds, val_ds, epochs, lr, model_name):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=['accuracy']
    )
    cb = [
        callbacks.ModelCheckpoint(os.path.join(MODEL_DIR, f'{model_name}.h5'), save_best_only=True, monitor='val_accuracy', mode='max'),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7),
        callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
    ]
    hist = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=cb)
    return hist

def fine_tune_model(model, base_model, train_ds, val_ds, unfreeze_fraction=0.3, epochs=10):
    base_model.trainable = True
    total = len(base_model.layers)
    fine_from = int(total * (1.0 - unfreeze_fraction))
    for layer in base_model.layers[:fine_from]:
        layer.trainable = False
    for layer in base_model.layers[fine_from:]:
        layer.trainable = True
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=['accuracy']
    )
    cb = [
        callbacks.ModelCheckpoint(os.path.join(MODEL_DIR, 'mobilenetv2_finetuned.h5'), save_best_only=True, monitor='val_accuracy'),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2),
        callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    ]
    hist = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=cb)
    return hist

# ---------------------------
# run CNN training
# ---------------------------
input_shape = IMG_SIZE + (3,)
print("\nBuilding MobileNetV2 CNN...")
cnn_model, cnn_base = build_mobilenetv2_cnn(input_shape, num_classes)

print("\nCNN Phase 1: feature extraction")
hist_cnn_phase1 = compile_and_train(cnn_model, train_ds, val_ds, epochs=EPOCHS_FEATURE, lr=1e-4, model_name='cnn_phase1')

print("\nCNN Phase 2: fine-tuning")
hist_cnn_ft = fine_tune_model(cnn_model, cnn_base, train_ds, val_ds, unfreeze_fraction=0.3, epochs=EPOCHS_FINETUNE)

print("\nEvaluate CNN on test set:")
cnn_loss, cnn_acc = cnn_model.evaluate(test_ds)
print(f"CNN Test accuracy: {cnn_acc*100:.2f}%")

# ---------------------------
# run hybrid (CNN+RNN)
# ---------------------------
train_hybrid = True   # set False to skip hybrid to save time
hybrid_model = None
if train_hybrid:
    print("\nBuilding CNN+RNN hybrid...")
    hybrid_model, hybrid_base = build_cnn_rnn_hybrid(input_shape, num_classes)
    hybrid_model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=['accuracy']
    )
    cb_h = [
        callbacks.ModelCheckpoint(os.path.join(MODEL_DIR, 'hybrid_phase1.h5'), save_best_only=True, monitor='val_accuracy'),
        callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6),
        callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
    ]
    print("\nHybrid Phase 1: feature extraction")
    hist_hybrid_p1 = hybrid_model.fit(train_ds, validation_data=val_ds, epochs=12, callbacks=cb_h)

    print("\nHybrid fine-tuning: unfreeze top CNN layers")
    hybrid_base.trainable = True
    for layer in hybrid_base.layers[:-40]:
        layer.trainable = False
    hybrid_model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
        metrics=['accuracy']
    )
    hist_hybrid_ft = hybrid_model.fit(train_ds, validation_data=val_ds, epochs=12, callbacks=cb_h)

    print("\nEvaluate Hybrid on test set:")
    h_loss, h_acc = hybrid_model.evaluate(test_ds)
    print(f"Hybrid Test accuracy: {h_acc*100:.2f}%")

# ---------------------------
# plots (optional)
# ---------------------------
def plot_history(history, title=None):
    if history is None: return
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='val')
    plt.title((title or '') + ' Loss')
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='val')
    plt.title((title or '') + ' Accuracy')
    plt.legend()
    plt.show()

try:
    plot_history(hist_cnn_phase1, 'CNN Phase1')
    plot_history(hist_cnn_ft, 'CNN Fine-tune')
    if train_hybrid:
        plot_history(hist_hybrid_p1, 'Hybrid Phase1')
        plot_history(hist_hybrid_ft, 'Hybrid Fine-tune')
except Exception:
    pass

# ---------------------------
# Grad-CAM + auto-misclassified
# ---------------------------
import matplotlib.cm as cm
from tensorflow.keras.models import Model

def find_last_conv_layer_name(model):
    candidates = []
    if hasattr(model, 'base_model'):
        candidates.append(model.base_model)
    candidates.append(model)
    for m in candidates:
        for layer in reversed(m.layers):
            if 'conv' in layer.name or (hasattr(layer.output, 'shape') and len(layer.output.shape) == 4):
                return layer.name
    raise RuntimeError("No conv layer found")

def get_img_array_from_path(path, size=IMG_SIZE):
    img_raw = tf.io.read_file(path)
    img = tf.image.decode_image(img_raw, channels=3, expand_animations=False)
    img = tf.image.resize(img, size)
    img = tf.cast(img, tf.float32) / 255.0
    return tf.expand_dims(img, axis=0)

def make_gradcam_heatmap(img_array, model, last_conv_name, pred_index=None):
    grad_model = Model(inputs=[model.inputs], outputs=[model.get_layer(last_conv_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array, training=False)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0].numpy()
    pooled_grads = pooled_grads.numpy()
    for i in range(pooled_grads.shape[-1]):
        conv_outputs[:, :, i] *= pooled_grads[i]
    heatmap = np.mean(conv_outputs, axis=-1)
    heatmap = np.maximum(heatmap, 0)
    maxv = np.max(heatmap) if np.max(heatmap) != 0 else 1e-10
    heatmap /= maxv
    return heatmap

def overlay_and_show(orig_img, heatmap, alpha=0.4, cmap='jet'):
    heatmap_resized = tf.image.resize(heatmap[..., np.newaxis], orig_img.shape[:2]).numpy().squeeze()
    col = cm.get_cmap(cmap)(heatmap_resized)[:, :, :3]
    overlay = orig_img * (1 - alpha) + col * alpha
    overlay = np.clip(overlay, 0, 1)
    return overlay

def grad_cam_visualize_for_paths(model, paths, top_k=2, save_dir=None):
    last_conv = find_last_conv_layer_name(model)
    os.makedirs(save_dir or './gradcam_out', exist_ok=True)
    for i, p in enumerate(paths):
        img_arr = get_img_array_from_path(p)
        preds = model.predict(img_arr)[0]
        top_idxs = np.argsort(preds)[::-1][:top_k]
        orig = img_arr[0].numpy()
        plt.figure(figsize=(4*(top_k+1),4))
        plt.subplot(1, top_k+1, 1); plt.imshow(orig); plt.title("Original"); plt.axis('off')
        for j, idx in enumerate(top_idxs):
            heat = make_gradcam_heatmap(img_arr, model, last_conv, pred_index=idx)
            overlay = overlay_and_show(orig, heat)
            plt.subplot(1, top_k+1, j+2)
            lbl = unique_labels[idx] if idx < len(unique_labels) else str(idx)
            plt.imshow(overlay); plt.title(f"{lbl} ({preds[idx]:.3f})"); plt.axis('off')
            if save_dir:
                outp = os.path.join(save_dir, f'gradcam_{i}_top{j}_cls{idx}.png')
                plt.imsave(outp, overlay)
        plt.show()

def pick_misclassified_samples(model, test_filepaths, test_labels, n=6):
    mis = []
    for p,y in zip(test_filepaths, test_labels):
        img = get_img_array_from_path(p)
        pred = np.argmax(model.predict(img)[0])
        if pred != y:
            mis.append((p,y,pred))
        if len(mis) >= n:
            break
    return mis

test_fp_list = list(test_fp)
test_labels_list = list(y_test)
best_for_viz = hybrid_model if (train_hybrid and hybrid_model is not None) else cnn_model

try:
    mis = pick_misclassified_samples(best_for_viz, test_fp_list, test_labels_list, n=6)
    if mis:
        print("Misclassified found; visualizing Grad-CAM on them...")
        grad_cam_visualize_for_paths(best_for_viz, [m[0] for m in mis], top_k=2, save_dir='./gradcam_auto')
    else:
        sample_paths = random.sample(test_fp_list, min(6, len(test_fp_list)))
        grad_cam_visualize_for_paths(best_for_viz, sample_paths, top_k=2, save_dir='./gradcam_auto')
except Exception as e:
    print("Grad-CAM failed:", e)
    print("Call grad_cam_visualize_for_paths(model, paths) manually")

# ---------------------------
# save label map
# ---------------------------
with open(os.path.join(MODEL_DIR, 'label_map.json'), 'w') as f:
    json.dump({'labels': unique_labels}, f, indent=2)

print("Done. Models saved to", MODEL_DIR)
print("Label map saved to", os.path.join(MODEL_DIR, 'label_map.json'))
```
